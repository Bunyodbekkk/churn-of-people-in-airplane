```python
x=df1.drop('satisfaction',axis=1)
y=df1['satisfaction']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain,ytrain,xtest,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn import metrics
ypred1=tree_model.predict(ytrain)
print(f"Model aniqligi: {metrics.accuracy_score(ytest,ypred1)*100} %")
```
